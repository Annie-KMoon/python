```python
#유사도 함수정리
def def_cosine_sim():
    #데이터 읽기
    import pandas as pd
    df = pd.read_csv('data/movie/tmdb_5000_movies.csv')
    df.fillna({'overview':''},inplace=True)  #널값데이터 전처리
    #BOW 생성
    from sklearn.feature_extraction.text import TfidfVectorizer
    tfidf = TfidfVectorizer(stop_words='english')
    tfidf_matrix = tfidf.fit_transform(df['overview'])
    #문장유사도
    from  sklearn.metrics.pairwise import linear_kernel
    cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

    return cosine_sim
```

```python
cosine_sim = def_cosine_sim()
cosine_sim
```

```python
#타이틀입력-유사영화탑10추천 함수
def recommend(title):
    import pandas as pd
    import pickle
    
    df = pd.read_csv('data/movie/tmdb_5000_movies.csv')
    idx = df[df['title']==title].index[0]

    cosine_sim = pickle.load(open('data/movie/cosine_sim.pickle','rb')) #파일 있을때 리드
    sim = cosine_sim[idx]
    
    sim = list(enumerate(sim))
    sim = sorted(sim, key=lambda x:x[1], reverse=True)
    sim = sim[1:11]
    index = [x[0] for x in sim]

    return index
```

```python
#장르, 키워드, 캐스트 상위 추출

def get_list(x):
    if isinstance(x, list): #X가 리스트배열일때
        names = [i['name'] for i in x]
        if len(names) >3:
            names = names[:3]
        return names
    return []
```

```python
def clean_data(x):
    if isinstance(x,list):
        return [str.lower(i).replace(' ','') for i in x]
    else:
        if isinstance(x,str):
            return str.lower(x).replace(' ','')
        return ''
```

```python
cols = ['director','cast', 'keywords', 'genres']
for col in cols:
    merge[col] = merge[col].apply(clean_data)
```

```python
def create_soup(x):
    str = x['director'] + ' '
    str += ' '.join(x['keywords'])+ ' '
    str += ' '.join(x['cast'])+ ' '
    str += ' '.join(x['genres'])+ ' '
    return str
```

```python
merge['soup'] = merge.apply(create_soup, axis=1)
```

```python
#유사도 함수2 감독/출연배우/장르로 유사도 연산
def def_cosine_sim2():
    #BOW 생성
    from sklearn.feature_extraction.text import CountVectorizer #카운터사용
    counter = CountVectorizer(stop_words='english')

    #데이터 읽기
    import pandas as pd
    df = pd.read_csv('data/movie/movies.csv')
    df.fillna({'soup':''},inplace=True)  #널값데이터 전처리

    #c_m생성
    counter_matrix = counter.fit_transform(df['soup'])
    
    #문장유사도
    from  sklearn.metrics.pairwise import linear_kernel
    cosine_sim2 = linear_kernel(counter_matrix, counter_matrix)

    return cosine_sim2
```

```python
cosine_sim2 = def_cosine_sim2()
cosine_sim2.shape
```

```python
import pickle
pickle.dump(cosine_sim2, open('data/movie/cosine_sim2.pickle','wb'))
```
